```python
# goto News Service
driver.get("https://news.naver.com/main/ranking/popularDay.nhn")
# crawling news article
import re 
p = "<dt>\s*?<a\shref=\"(.*?)\".*?>(.*?)<\/a>"
article_p = '<em\sclass=\"img_desc\">(.*?)<a\shref.*?>'
temp = re.findall(p,driver.page_source)

scrapedNews = []
for news in temp: 
    print(news[1])
    try:
        driver.get("https://news.naver.com"+news[0].replace("&amp;","&"))
        article = re.findall(article_p ,driver.page_source)
        article = re.sub("<.*?>","",article[0])
        scrapedNews.append({
            'title':news[1],
            'article' :article 
        })
    except:
        print("article Loading Fail")
```
