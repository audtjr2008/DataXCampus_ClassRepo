## 13일차 과제수행 이미지 크롤링, MongoDB 저장- 김명석

과제는 이미지를 크롤링 하는것.

리퀘스트한 파일을 utf8 로 디코딩 해야하고, base64 이미지파일을 받아오면 된다.
과제는 이미지 크롤링 한다.
르루아르, 샤갈, 모딜리아니 작품들을 크롤링한다.
작품들 클러스터링하기
blob 사용해보기, base64.ㅡ

### Request 

```python
import re
import requests
def getImgFromGoogle(keyword):
    url = 'https://www.google.com/search'
    def_params = {
        'q': None,
        "source":"lnms",
        'tbm':'isch',
        'sa':'X',
        "biw":1536
    }
    def_params['q'] = keyword
    p = '<img.*?src="(.*?)".*?>'
    
    resp = requests.get(url,params=def_params)
    return re.findall(p,resp.text)[1:]


```

```python
list(getImgFromGoogle("르누아르"))
```

```python
from pymongo import MongoClient
client = MongoClient("localhost")

imageDB = client.get_database("imageDB")
try:
    imageBase64 = imageDB.create_collection("imageBase64")
except:
    imageBase64 = imageDB.get_collection("imageBase64")

def insert_img(pName , pBase64):
    file = {
    'name' : None,
    'base64':None,
    }
    file['name'] = pName
    file['base64'] = pBase64
    return imageBase64.insert_one(file)
```

```python
import urllib.request 
import base64
import os
keyword = "르누아르"
files=[]
for i, url in enumerate(getImgFromGoogle(keyword)):
    files.append(urllib.request.urlretrieve(url,keyword+str(i)+".jpg")[0])

for file in files:
    with open(file,'rb') as fd:    
        aBase64 = base64.encodebytes(fd.read())
        print(insert_img(file,aBase64))
    os.remove(file)

```
